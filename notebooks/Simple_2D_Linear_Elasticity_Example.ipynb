#Linear Elasticity Simulations and the Materials Knowledge System

##Introduction

This notebook provides an example of using the Materials Knowledge System (MKS) to compute the local strain field of a composite material. The notebook first provides some background information on linear elasticity as well as the boundary conditions of the example. Sfepy was used to compute the calibration and the validation datasets. The notebook then demonstrates the functions and processes need to calibrate and validate the MKS frame for a material system.

###Elastostatics Equations

The constitutive equation that describes the linear elastic phenomena is Hook's law.

$$ \sigma_{ij} = C_{ijkl}\varepsilon_{kl} $$

$\sigma$ is the stress, $C$ is the stiffness tensor, and $\varepsilon$ is the strain. For an isotropic material the stiffness tensor can be that relates the stress and the strain as follows.

$$ \sigma_{ij} = \lambda \delta_{ij} \varepsilon{kk} + 2\mu \varepsilon_{ij}  $$

$ \lambda$ and $\mu$ are the first and second lam$\acute{e}$ parameters and can be defined in terms of the Young's modulus $E$ and Poisson's ratio $\nu$.

$$ \lambda = \frac{E\nu}{(1-\nu)(1-2\nu)} $$

$$ \mu = \frac{E}{2(1+\nu)} $$


Strain is related to displacement using the following equation.

$$ \varepsilon_{ij} = \frac{u_{i,j}+u_{j,i}}{2} $$

Plugging the equation above back into our expression for stress, we get the following equation.

$$ \sigma_{ij} = \lambda u_{k,k} + \mu( u_{i,j}+u_{j,i})  $$

Substituting the divergence of the stress into the equilibrium equations allows for the displacements to be solved instead of stress.

$$ \lambda u_{k,k} + \mu( u_{i,j}+u_{j,i}) + F_i = 0 $$

In this example, a displacement controlled simulation is used to calculated the strain. The body force $F_i$ is set equal to zero

$$ \mu u_{i,jj}+(\mu + \lambda)u_{j,ij}=0 $$

and the following boundary conditions for square box of length $L$ and macroscopic strain $\bar{\varepsilon}$ are imposed.

$$ u(x,L) = u(x,0)   $$
$$ u(0,y) = 0 $$
$$ u(L,y) = L\bar{\varepsilon} $$



###Sfepy - Simple Finite Elements in Python

In this example the an instance of the class `ElasticFEModel` is used to generate both the calibration strain response field to fit the MKS model and the validation strain response field used to evaluate the model's accuracy. The class utilizes the module Sfepy to calculated the local response strain field though finite element methods. More information can be found their [website](http://sfepy.org/doc-devel/index.html).

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
```

##Simulation

###Calibration Data and Delta Microstructures

It has been shown that the first order MKS influence coefficients are sufficient to compute a strain field of a random microstructure with a contrast of less than 1.5 with a mean absolute error on the order of 2% relative to the strain field computed using finite element methods [[1]](#ref_1). 

The first order coefficients can be efficiently calibrated using delta microstructures which are analogous to unit impulses in signal processing [[2]](#ref_2). Delta microstructures are composed of only two phases. One phase can be found throughout the entire microstructure except at the very center which contains a different phase. For this example (and any two phase system) only two delta microstructures are needed to capture the first order coefficients. 

###Generating Calibration Data

Define an array that contains the Young's modulus `E` and Poission's ratio `nu` for 2 `L` by `L` delta microstructures.


What units does Sfepy expect the Young's modulus to have?

```python
E1 = 80 #Young's modules for material 1
E2 = 120 #Young's modules for material 2
nu1 = 0.3 #Poisson's ratio for material 1
nu2 = nu1 #Poisson's ratio for material 2

L = 21
C = (L-1)/2

Delta_prop = np.zeros((2, L, L, 2))
Delta_prop[...] = E1
Delta_prop[:, C, C, :] = E2
Delta_prop[1] = np.where(Delta_prop[1] == E1, E2, E1)
Delta_prop[..., 1] = nu1
Delta_prop[:, C, C, 1] = nu2
Delta_prop[1, :, :, 1] = np.where(Delta_prop[1, :, :, 1] == nu1, nu2, nu1)

plt.imshow(Delta_prop[1,:,:,0])
plt.title('Delta Microstructure',fontsize=20)
```

Strain field responses must be computed for the delta microstructures. This can be done by importing `ElasticFEModel` from `pymks`, or though another FE tool of your choice.

```python
from pymks import ElasticFEModel

FEmodel = ElasticFEModel()
y = FEmodel.predict(Delta_prop)
```

Let's take a look that the $\varepsilon_{xx}$, $\varepsilon_{yy}$, and $\varepsilon_{xy}$ strain fields produced by the `FEmodel`.

```python
p = ['xx','yy','xy']
for ii in range(3):
    plt.figure()
    plt.title('FE Calculated '+p[ii]+' Strain Field',fontsize=15)
    plt.imshow(y[0, :, :, ii])
    plt.colorbar()
```

###Calibrating First Order Influence Coefficients

Create an instance of MKSRegressionModel and create the microstructure function.

```python
from pymks import MKSRegressionModel
MKSmodel = MKSRegressionModel(Nbin = 2)

X_MSf = np.zeros((2, L, L))
X_MSf[0, C, C] = 1
X_MSf[1] = np.abs(X_MSf[0]-1)
```

The rest of this example will only use the $\varepsilon_{xx}$ strain field, but the others can be calculated using obvious substitutions.

The first order influence coefficient can be created by passing the calculated strain field from the FE simulation and the microstructure function as arguments into the `fit()` method.

```python
MKSmodel.fit(X_MSf,y[..., 0])
```

Let's take a look at the influence coefficients.

```python
for ii in range(2):
    plt.figure()
    plt.imshow(MKSmodel.coeff[...,ii])
    plt.title('Influence Coefficients',fontsize=20)
    plt.colorbar()
```

The first set of influence coefficients have a Gaussian shape to them while the second influence coefficients are a constant. The constant influence coefficients may seem superfluous, but are analogous to the constant value found doing multiple linear regression with categorical variables.

### Validation of the Calibrated Influence Coefficients

Now let's use the calibrated influence coefficients to compute the local strain response for a random two phase microstructure. First create a random microstructure.

```python
np.random.seed(99)
RM = np.random.random((1, L, L))
RM = np.where(RM > 0.5, 0, 1)
plt.imshow(RM[0])
plt.title('Random Microstructure',fontsize=18)
```

Create a property array that assigns the Young's modulus `E` and the Possion's ratio `nu` for both of the phases.

```python
RM_prop = np.zeros((1, L, L, 2))
RM_prop[..., 0] = np.where(RM == 0, E1, E2)
RM_prop[..., 1] = np.where(RM == 0,nu1, nu2)
```

Create the microstructure function.

```python
RM_MSf = np.where(RM[0] == 1, 1, 0)
RM_MSf[1] = np.abs(RM_MSf[0] - 1)
```

Compute the local strain field for the random microstructure using the FE model.

```python
y = FEmodel.predict(RM_prop)
plt.imshow(y[0, :, :, 0])
plt.title('FE Computed Local Strain Field',fontsize=15)
plt.colorbar()
```

The local strain field can be computed by passing the random microstructure as an argument to the `predict()` method of the MKS model.

```python
yMKS = MKSmodel.predict(RM)
plt.imshow(yMKS[0])
plt.title('MKS Computed Local Strain Field',fontsize=15)
plt.colorbar()
```

Calculate the mean absolute error.

```python
print np.average(np.abs(yMKS[0]-y[0,:,:,0])/y[0,:,:,0])*100
```
