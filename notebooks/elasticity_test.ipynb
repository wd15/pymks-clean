```python
from pymks import ElasticFEModel

E0 = 1.1
E1 = 1.
nu = 0.3

nx = 11
Lx = 1.
dx = Lx / nx
ii = (nx - 1) / 2


X = np.zeros((2, nx, nx))
X_prop = np.zeros(X.shape + (2,))
X[0, ii, ii] = 1.
X[1] = -X[0] + 1.


X_prop[...,1] = nu
for i in (0, 1):
    X_prop[i,...,0] = E1 * X[i] + E0 * (1 - X[i])

elastic_model = ElasticFEModel(dx=dx)
y_prop = elastic_model.predict(X_prop)
```

```python
from pymks import MKSRegressionModel

y_test = y_prop[...,0]

model = MKSRegressionModel(Nbin=2)
model.fit(X, y_test)
```

```python
y_pred = model.predict(X)
```

```python
plt.imshow(y_pred[0])
plt.colorbar()
for i in (5, 6, 7):
    print y_pred[0,i]
```
