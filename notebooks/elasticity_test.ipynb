```python
from pymks import ElasticFEModel

E0 = 1.1
E1 = 1.
nu = 0.3

nx = 21
ny = 21
dx = 1.
ii = (nx - 1) / 2
jj = (ny - 1) / 2


X = np.zeros((2, nx, ny))
X_prop = np.zeros(X.shape + (2,))
X[0, ii, jj] = 1.
X[1] = -X[0] + 1.


X_prop[...,1] = nu
for i in (0, 1):
    X_prop[i,...,0] = E1 * X[i] + E0 * (1 - X[i])

elastic_model = ElasticFEModel(dx=dx)
y_prop = elastic_model.predict(X_prop)
```

```python
from pymks import MKSRegressionModel

model = MKSRegressionModel(Nbin=2)
model.fit(X, y_prop[...,0])
```

```python
coeff = np.fft.ifftn(model.Fcoeff, axes=(0, 1))
```

```python
coeff.shape
```

```python
plt.contourf(np.fft.fftshift(coeff[...,0]))
plt.colorbar()
```

```python
plt.contourf(np.fft.fftshift(coeff[...,1]))
plt.colorbar()
```
