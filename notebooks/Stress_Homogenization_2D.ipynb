#Effective Stress Homogenization Linkage using MKSHomogenzationModel

This goal of this example is to create an efficient homogenization linkage for stress values given a random set of microstructures.

Let first start by importing a few functions that we will need through out the rest of this example.

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
```

## Generate Data

A set of microstructures and their efficient stress values can be generated by importing `make_elastic_stress_random` function from `pymks.datasets`. In this function we will us the key work argument `n_sample` to specify how many samples we need, `grain_size` to give the effective microstructure feature size. The arguments `elastic_modulus` and `poissons_ratio` are used to give the material property for each of the phases and the number of values in each of these argument indicates the number of phases. Lastly, we will using the `macro_strain` to indicate the value of the macroscopic strain applied in the x-direction for each of the microstructures and the variable `seed` can be used to change the seed of the random number generator.

```python
from pymks.datasets import make_elastic_stress_random

X, y = make_elastic_stress_random(n_samples=400, grain_size=(15, 2), elastic_modulus=(500, 100), 
                                  size=(21, 21), poissons_ratio=(1, 3), macro_strain=1., seed=4)
print X.shape
print y.shape
```

The array `X` contains the microstructure information and has the dimensions of `(n_samples, Nx, Ny)`. The array `y` contains the effective stress information and has dimensions of `(n_samples,)`.

Lets take a look at 3 of the microstructures to get an idea of what they look like by importing `draw_microstructures` from `pymks.tools`.

```python
from pymks.tools import draw_microstructures

draw_microstructures(X[:3])
```

Let also take a look at their assoicated stress values.

```python
print y[:3]
```

## Specify Dimensionality Reduction and Property Linkage Techniques

In order to make instantiate an instance of a MKSHomogenizationModel, we need to specify the dimensionality reduction technique and the property linking method that we want to use.

### Dimensionality Reduction Technique

For this example we are going to use Singular Value Decomposition (SVD) for our dimensionality reduction technique. Because the property we are looking to predict is linear, we could have chosen, PCA or randomized PCA as well. All these methods are similar, but SVD is more computationally efficient.

Let's make an instance of [TruncatedSVD](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html) from [sklearn](http://scikit-learn.org/stable/) that we will use to make our MKSHomogenization model. We will set a large number of components (number of eigenvectors) to evaluate the amount of the data
variance then we will use.

```python
from sklearn.decomposition import TruncatedSVD

dimension_reducer = TruncatedSVD(n_components=20)
```

### Property Linkage Techinque

At has been shown that linear regression with polynomial gives the best results (add ref). We will using [Pipline](http://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html) from sklearn to create a polynomial version of the linear regresison model using [PolynomialFeatures](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html) and [LinearRegression](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html). We start with a second order polynomial.

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

property_linker = Pipeline([('poly', PolynomialFeatures(degree=1)),
                  ('linear', LinearRegression())])
```

## Create an Instance of MKSHomogenizationModel

Now that we have created instances of our dimensionality reduction and property linkage techniques we almost ready to create an instance of the MKSHomogenizationModel. The last decision we need to make is how we are going to discretize our microstructures. Because we are using discrete phases, the `DiscreteIndicatorBasis` is appropreate. For all bases in pymks the number of local states (in this chase the number of phases) and their domain need to be specified. We have havd an instance of our basis we can now create an instance of the MKSHomogenizationModel. 

```python
from pymks import MKSHomogenizationModel, DiscreteIndicatorBasis

basis = DiscreteIndicatorBasis(2, [0, 1])
model = MKSHomogenizationModel(basis=basis, dimension_reducer=dimension_reducer, property_linker=property_linker)
```

Now that we have our model, let's look to see if we have a sufficient amount of data to predict the effective stress. We can do this by using `draw_learning_curves` from `pymks.tools`. We will provide the it with an instance of our model and our data (X and y). The parameter `n_jobs` can be used to state the number of jobs to be run in parallel.

```python
from pymks.tools import draw_learning_curves

draw_learning_curves(model, X, y, n_jobs=6)
```

Because the train score and cross validation scores start to converge it appears that we have enough data to work with.

Let's take a look at how the variance changes as a function of component. To do this we will import `draw_component_variance` from `pymks.tools`. For most of the dimensionality reduction classes form sklearn have a
`explained_variance_ratio_` method that can be used, but the documentation for each method should checked.

```python
from pymks.tools import draw_component_variance

model.fit(X, y)
variance = model.dimension_reducer.explained_variance_ratio_
draw_component_variance(variance)
```

Although the variance does not seem to change much over after 10 components, the small variation in one of components make have
a large impact on the fit of our model.

Now lets looks at how our model changes as we increase the order of our polynomial terms in the linear regression modeol. To do this, we will first slipt the data into a testing and training sets using `train_test_spilt` from `sklearn`.

```python
from sklearn.cross_validation import train_test_split

flat_shape = (X.shape[0],) + (np.prod(X.shape[1:]),)

X_train, X_test, y_train, y_test = train_test_split(X.reshape(flat_shape), y,
                                                    test_size=0.8, random_state=3)
print X_train.shape
print X_test.shape
```

Now that we have our testing and training datasets, use the 

```python
from sklearn.metrics import make_scorer, r2_score, mean_squared_error
from sklearn.grid_search import GridSearchCV

model.dimension_reducer.n_components = 10

params_to_tune = {'property_linker__poly__degree': np.arange(1, 7)}
fit_params = {'size': X[0].shape}
gs = GridSearchCV(model, params_to_tune, cv=3, n_jobs=6, fit_params=fit_params).fit(X_train, y_train)
```

```python
print(gs.best_estimator_)
print(gs.score(X_test, y_test))
```

```python
from pymks.tools import draw_gridscores

draw_gridscores(gs.grid_scores_, 'Truncated SVD', '#1a9641', parameter='property_linker__poly__degree',
                axis_label='Polynomial Degree')
legend = plt.legend()
```

```python
n_test_samples = 25

property_linker = Pipeline([('poly', PolynomialFeatures(degree=1)),
                            ('linear', LinearRegression())])
dimension_reducer = TruncatedSVD(n_components=15)

model = MKSHomogenizationModel(basis=basis, property_linker=property_linker, dimension_reducer=dimension_reducer)
model.fit(X[:-n_test_samples], y[:-n_test_samples])
```

```python
from pymks.tools import draw_goodness_of_fit


draw_goodness_of_fit(y[:-n_test_samples], model.predict(X[:-n_test_samples]))
```

```python
y_predict = model.predict(X[-n_test_samples:])
```

```python
from pymks.tools import draw_components

draw_components(model.fit_data[:, :2], model.predict_data[:, :2],
                label1='Fit Data', label2='Predict Data')
```

```python
print r2_score(y[-n_test_samples:], y_predict)
```
