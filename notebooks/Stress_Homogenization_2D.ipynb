#Effective Stress Homogenization Linkage using MKSHomogenzationModel

This example uses the MKSHomogenizationModel to create a homogenization linkage for 
average stress. This example start with some background theory on effective elastic 
modulus of a composite materials. Microstructures and their average stress values are
generated for the model. Model selection for MKSHomogenizationModel parameters are
chosen and the model accurately predicts the stress values for a new set of microstructures.

## Linear Elasticity and Effective Elastic Modulus

For this homogenization linkage, we are looking to predict the average isotropic stress on a 
random microstructure when a uniaxial strain is applied at the boundaries. The average
stress $\bar{\sigma}$ and average strain $\bar{\varepsilon}$ values can be related to each other 
through the effective elastic modulus $C^*$. 

$$ \bar{\sigma}_{ij} = C_{ijkl}^* \bar{\varepsilon}_{kl} $$

The [rule of mixtures](http://en.wikipedia.org/wiki/Rule_of_mixtures) provides simple 
analytical equations that provide an upper and lower bound on an effective property, 
but the solutions are limited to very unique microstructures. Effective properties 
are difficult to compute because they generally cannot be computed using a simple average.

$$ \bar{C}_{ijkl} \neq C_{ijkl}^* $$

In this example, we generate a set of microstructures and compute their local strain values
at each localtion using the following boundary condition.

$$ u(L, y) = u(0, y) + L\bar{\varepsilon}_{xx}$$

$$ u(0, L) = u(0, 0) = 0  $$

$$ u(x, 0) = u(x, L) $$

The local strain values are then multiplied by the local elastic modulus to get the local 
strain values. The strain values are then averaged.

## Generate Data

A set of periodic microstructures and their efficient stress values can be generated by 
importing `make_elastic_stress_random` function from `pymks.datasets`. In this 
function the argument `n_samples` is the number of samples, `size` 
specifices the number of voxels as well as the dimensions of the microstructures,
`grain_size` controls the effective microstructure feature size, `elastic_modulus` 
and `poissons_ratio` are used to indicate the material property for each of the 
phases, `macro_strain` is the value of the applied uniaxixial strain, and the 
variable `seed` can be used to change the the random number generator seed. 

Let's go ahead and create 6 sets of 200 samples of 21 x 21 microstructures with differnt
effective microstructure feature sizes and their associated stress values.

```python
from pymks.datasets import make_elastic_stress_random
sample_size = 200
grain_size = [(15, 2), (2, 15), (7, 7), (8, 3), (3, 9), (2, 2)]
n_samples = [sample_size, sample_size, sample_size,
             sample_size, sample_size, sample_size]
elastic_modulus = (410, 200)
poissons_ratio = (0.28, 0.3)
macro_strain = 0.001
size = (21, 21)

X, y = make_elastic_stress_random(n_samples=n_samples, size=size, grain_size=grain_size, 
                                      elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio, 
                                      macro_strain=macro_strain, seed=0)
```

The array `X` contains the microstructure information and has the dimensions 
of `(n_samples, Nx, Ny)`. The array `y` contains the average stress value for 
each of the microstructures and has dimensions of `(n_samples,)`.

```python
print(X.shape)
print(y.shape)
```

Lets take a look at the 6 types the microstructures to get an idea of what they 
look like by importing `draw_microstructures` from `pymks.tools`. 

```python
from pymks.tools import draw_microstructures
X_examples = X[::sample_size]
draw_microstructures((X_examples[:3]))
```

```python
draw_microstructures((X_examples[3:]))
```

The dataset has 4 types of examples that grains that are elongated in either
the x or y directions. The remaining 2 types of samples are equiaxial with
different grain sizes.

Let's look at the stress values for each of the mictrosturtures shown above.

```python
print('Stress Values'), (y[::200])
```

Now that we have a dataset to work with, we can look at how to use the `MKSHomogenizationModel`to predict stress values for new microstructures.

## MKSHomogenizationModel Work Flow

The default instance of the MKSHomogenizationModel takes in a dataset and 
 - calculates the 2-point statistics
 - performs [dimensionality reduction](http://en.wikipedia.org/wiki/Dimensionality_reduction) using [Singular Valued Decomposition](http://en.wikipedia.org/wiki/Singular_value_decomposition) (SVD)
 - and fits a [polynomial regression model](http://en.wikipedia.org/wiki/Polynomial_regression) model to the low-dimensional representation.

These work flow has been shown to accurately predict effective properties in several examples [1][2].

#Modeling with MKSHomogenizationModel

In order to make an instance of the MKSHomogenizationModel, we need to pass instance of a basis (used to compute the 2-point statistics). For this particular example, there are only 2 discrete phases, so we will use the `DiscreteIndicatorBasis` from `pymks`.

```python
from pymks import MKSHomogenizationModel, DiscreteIndicatorBasis

basis = DiscreteIndicatorBasis(n_states=2, domain=[0, 1])
model = MKSHomogenizationModel(basis=basis)
```

Now we have a MKSHomogenizationModel. Let look at the default values for the number of components and the order of the polynomial.

```python
print('Default Number of Components'), (model.n_components)
print('Default Polynomail Order'), (model.poly_order)
```

These default parameters may not be the best model for a given problem. Let's look at how we can optmize them.

## Optimizing the Number of Components and Polynomial Order

To start with, let's take a look at how the variance changes as a function of the number of components.
In general for SVD as well as PCA, the 1st component captures the most variance in the data compared
to the other components, the second component captures the second most variance, and so on.
This means that the number of components increases, the percentage of the variance that is
captured should also increase. Let's see if this is true for our dataset.

In order to do this we will change the number of components from 2 to 40 and then
fit the data we have. The `fit` function performs the dimensionality reduction and 
also fits the regression model. Because our microstructures are periodic, we need to 
use the `periodic_axes` argument when we `fit` the data.

```python
model.n_components = 40
model.fit(X, y, periodic_axes=[0, 1])
```

Let's take a look how variance changes by importing `draw_component_variance` 
from `pymks.tools`.

```python
from pymks.tools import draw_component_variance

draw_component_variance(model.dimension_reducer.explained_variance_ratio_)
```

90 percent of the variance is captured with the first 5 components. This means our model will likely only need a few components to
predict the average stress.

Next we need to optimze the number of components and the polynomial order. Do do this we are going to split the data into testing and training sets. This can be done using using [train_test_spilt](http://scikit-learn.org/stable/modules/generated/sklearn.cross_validation.train_test_split.html) function from sklearn.

```python
from sklearn.cross_validation import train_test_split

flat_shape = (X.shape[0],) + (np.prod(X.shape[1:]),)

X_train, X_test, y_train, y_test = train_test_split(X.reshape(flat_shape), y,
                                                    test_size=0.8, random_state=3)
print(X_train.shape)
print(X_test.shape)
```

Now that we have our testing and training datasets, we will fit the testing data to a number of models using cross validation
while varying number of components and polynomail orders, and then use the test data to verify the model. This can be done using [GridSeachCV](http://scikit-learn.org/stable/modules/generated/sklearn.grid_search.GridSearchCV.html) from sklearn.

In order to do this we will pass a dictionary `params_to_tune` with the range of
polynomial order `poly_order` and components `n_components` we want to try.
A dictionary `fit_params` can be used to pass the `periodic_axes` variable to 
calculate periodic 2-point statistics. The argument `cv` can be used to specify 
the number of folds used in crossvalidation and `n_jobs` can be used to specify 
the number of jobs that are ran in parallel.

Let's vary both `n_components` from 1 to 4 and `poly_order` from 1 to 3.

```python
from sklearn.grid_search import GridSearchCV

params_to_tune = {'poly_order': np.arange(1, 4), 'n_components': np.arange(1, 7)}
fit_params = {'size': X[0].shape, 'periodic_axes': [0, 1]}
gs = GridSearchCV(model, params_to_tune, cv=12, n_jobs=6, fit_params=fit_params).fit(X_train, y_train)
```

The default `score` method for the is R-squared value. Let's look at the how the mean values of R-squared and their 
strandard deviations change as we varied the number of `n_components` and `poly_order`. We can do that by importing
`draw_gridscores_matrix` from `pymks.tools`.

```python
from pymks.tools import draw_gridscores_matrix

draw_gridscores_matrix(gs, ['n_components', 'poly_order'], score_label='R-Squared',
                       param_labels=['Number of Components', 'Order of Polynomial'])
```

It looks like we get a poor fit with only the first and second component, and we we increase
the polynomial order to 4, the values look have a high standard deviation and poor score values.

It's difficult to see which model is the best. Let's use our testing data `X_test` to see which
model performs the best.

```python
print('Order of Polynomial'), (gs.best_estimator_.poly_order)
print('Number of Components'), (gs.best_estimator_.n_components)
print('R-squared Value'), (gs.score(X_test, y_test))
```

For the parameter range that we searched, we have found that a model with 3rd order polynomial 
and 3 components had the best R-squared value. It's difficult to see the differences in the score
values and the standard deviation when we have 3 or more components and a polynomial order of less
than 4. Let's take a closer look at those values using `draw_grid_scores`.

```python
from pymks.tools import draw_gridscores

gs_po_1 = [x for x in gs.grid_scores_ \
            if x.parameters['poly_order'] == 1][2:]
gs_po_2 = [x for x in gs.grid_scores_ \
            if x.parameters['poly_order'] == 2][2:]
gs_po_3 = [x for x in gs.grid_scores_ \
            if x.parameters['poly_order'] == 3][2:]

draw_gridscores(gs_po_3, 'n_components', data_label='3rd Order', color='#762a83',
                axis_label='Number of Components', score_label='R-Squared')
draw_gridscores(gs_po_2, 'n_components', data_label='2nd Order', color='#1a9641',
                axis_label='Number of Components', score_label='R-Squared')
draw_gridscores(gs_po_1, 'n_components', data_label='1st Order', color='#f46d43',
                axis_label='Number of Components', score_label='R-Squared')
```

It looks like a model with a 3rd order polynomial and 3 components will give us the best result,
but were are several other models that will likely provide comparable results. So, let's make the
best model from our grid searches.

```python
model = gs.best_estimator_
```

#Prediction using MKSHomogenizationModel

Now that we have selected a model, lets fit the model witht the data. Again because
our microstructures are periodic, we need to use the `periodic_axes` argument.

```python
model.fit(X, y, periodic_axes=[0, 1])
```

Let's generate some more data to try and predict with our model. We are going to
generate 20 samples of all six different types of microstructures.

```python
test_sample_size = 20
n_samples = [test_sample_size] * 6
X_new, y_new = make_elastic_stress_random(n_samples=n_samples, size=size, grain_size=grain_size, 
                                          elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio, 
                                          macro_strain=macro_strain, seed=1)
```

Now let's predict the stress values for the new microstructures. 

```python
y_predict = model.predict(X_new, periodic_axes=[0, 1])
```

We can look to see if the new data is similar to the data
we used to fit the model by looking at their low dimensional
representations using `draw_components` from `pymks.tools`.

```python
from pymks.tools import draw_components

draw_components(model.fit_data[:, :2], model.predict_data[:, :2],
                label1='Train Data', label2='Test Data')
```

The predicted data seems to be reasonable simlar to the data we fit the model
with. Now let's look at the score value for the predicted data.

```python
from sklearn.metrics import r2_score
print('R-squared'), (model.score(X_new, y_new, periodic_axes=[0, 1]))
```

Looks pretty good. Let's print out a few of the actual and predicted stress values to see
how they compare.

```python
print('Actual Stress   '), (y_new[::20])
print('Predicted Stress'), (y_predict[::20])
```

We can also evaluate our prediction by looking at a goodness-of-fit plot. We
can do this by importing `draw_goodness_of_fit` from `pymks.tools`.

```python
from pymks.tools import draw_goodness_of_fit

fit_data = np.array([y, model.predict(X)])
pred_data = np.array([y_new, y_predict])
draw_goodness_of_fit(fit_data, pred_data)
```

We can see that the MKSHomogenizationModel as predicted the average stress values for our new microstructures reasonably well.
