#Cahn-Hilliard 

This example demonstrates how to use PyMKS to quickly produce phase field results from a Cahn-Hilliard simulation. 

###Cahn-Hilliard Equation

$$ \frac{\partial \phi}{\partial t} = \nabla \cdot D \nabla \Bigg ( \frac{\partial f}{\partial \phi} + \epsilon \nabla^2 \phi \Bigg )$$

where $\phi$ is an conserved ordered parameter can represent alloy composition. The double well free energy function $f$ is parameterized by

$$ f = \frac{a^2}{2}\phi^2 (1 - \phi^2)$$

where the free paramter $a$ can be used to control the high of the free energy function. The double well provide the thermodynamic driving for for phase seperation. While the gradient energy term $\epsilon \nabla^2 \phi$ provides the driving force to keep the interface between the phases continous. The free parameter $ \epsilon$ can be used to control to interface distance.



###Calibration Datasets

The MKS coefficients can be calibrated using regression.

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
```

```python
import pymks
from pymks.datasets import make_cahnHilliard

L = 21
Nsamples = 4000
dt = 1e-4
dx = 1.
width = 4.


X, y = make_cahnHilliard(n_samples=Nsamples, size=(L, L), dt=dt, dx=dx, width=width)
```

```python
from pymks.tools import draw_concentrations

draw_concentrations(X[0], y[0])
```

```python
plt.imshow(X[0] - y[0])
plt.colorbar()
```

### Calibrate Influence Coefficients

```python
from pymks import MKSRegressionModel
MKSmodel = MKSRegressionModel(Nstate=4)

MKSmodel.fit(X, y)
```

```python
from pymks.tools import draw_coeff

coeff = MKSmodel.coeff.real
draw_coeff(coeff[...,:])

```

```python
from IPython.display import clear_output
import time
from pymks.datasets.cahnHilliardSimulation import CahnHilliardSimulation
from pymks.tools import draw_concentrations_compare
np.random.seed(99)
N = 3 * L
phi0 = np.random.random((1, N, N))

steps = 300

MKSmodel.resize_coeff((N, N))

CHSim = CahnHilliardSimulation(dx=dx, dt=dt, width=width)
CHSim_response = phi0.copy()
MKS_response = phi0.copy()

for i in range(steps):
    CHSim_temp = CHSim.get_response(CHSim_response)
    CHSim_response = CHSim_temp
    MKS_temp = MKSmodel.predict(MKS_response)
    MKS_response = MKS_temp

draw_concentrations_compare(CHSim_response[0], MKS_response[0])
```
