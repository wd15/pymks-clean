```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pylab as plt
from pymks.datasets.microstructureGenerator import MicrostructureGenerator
```

```python
n_phases = 3
n_samples = 50
N = 100
X0 = MicrostructureGenerator(n_samples, (N, N), n_phases=n_phases, grain_size=(20,20)).generate()
X1 = MicrostructureGenerator(n_samples, (N, N), n_phases=n_phases, grain_size=(50,5)).generate()
```

```python

plt.imshow(X0[0])
plt.figure()
plt.imshow(X1[1])
plt.colorbar()
```

```python
from pymks.bases import DiscreteIndicatorBasis

basis = DiscreteIndicatorBasis(n_states=n_phases)
X0_ = basis.discretize(X0)
X1_ = basis.discretize(X1)
```

```python
from pymks.stats import autocorrelate
from pymks.stats import crosscorrelate

X0_auto = autocorrelate(X0_)
X1_auto = autocorrelate(X1_)
X0_cross = crosscorrelate(X0_)
X1_cross = crosscorrelate(X1_)

print X0_cross.shape
```

```python
plt.imshow(X0_auto[0,...,2])
plt.figure()
plt.imshow(X1_auto[0,...,2])
plt.colorbar()
```

```python
X0_pca = np.concatenate((X0_auto, X0_cross), axis=-1)
X1_pca = np.concatenate((X1_auto, X1_cross), axis=-1)


X0_pca = X0_pca.swapaxes(1, -1).reshape((n_samples, N * N * 2* n_phases))
X1_pca = X1_pca.swapaxes(1, -1).reshape((n_samples, N * N * 2* n_phases))
X_pca = np.concatenate((X0_pca, X1_pca))


X0_pca_tmp =np.concatenate((X0_auto, X0_cross))
X0_pca_tmp = X0_pca_tmp.swapaxes(1, -1).reshape((100, N * N * n_phases))
print X_pca.shape
print X0_pca.shape
print X0_pca_tmp.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_r = pca.fit(X0_pca_tmp).transform(X0_pca_tmp)
PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_) 

#print X0_pca.shape
```
