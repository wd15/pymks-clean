```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
```

```python
from pymks.datasets import make_microstructure

X = make_microstructure(n_samples=1, size=(101, 201))
```

```python
from pymks import DiscreteIndicatorBasis
from pymks.stats import autocorrelate

dbasis = DiscreteIndicatorBasis(2, [0, 1])
X_ = dbasis.discretize(X)
X_auto = autocorrelate(X_)
```

```python
from pymks.tools import draw_autocorrelations

draw_autocorrelations(X_auto[0])
```

```python
from pymks.tools import draw_microstructures

draw_microstructures(X)
```
