```python
import fipy as fp
```

```python
L = 1.0
N = 10
dx = L / N

shear_modulus = 1.0
lame = 1.0

f_x = 0.02
f_y = 0.01
```

```python
mesh = fp.Grid2D(nx=N, ny=N, dx=dx, dy=dx)

#mesh = fp.Grid2D(nx=N, ny=N, dx=dx, dy=dx)

#lame = fp.CellVariable(mesh=mesh, value=1.0)
#lame.setValue(10.0, where=mesh.x > 0.5)

u_x = fp.CellVariable(mesh=mesh)
u_y = fp.CellVariable(mesh=mesh)

u_x.constrain(0, mesh.facesLeft)
u_y.constrain(0, mesh.facesLeft)
u_x.constrain(0.01 * (mesh.y.faceValue - 0.5)**2, mesh.facesRight)
u_y.grad.constrain(0, mesh.facesRight)
#u_y.constrain(0, mesh.facesLeft)
```

```python
A = fp.CellVariable(mesh=mesh, rank=2)
B = fp.CellVariable(mesh=mesh, rank=2)
C = fp.CellVariable(mesh=mesh, rank=2)
D = fp.CellVariable(mesh=mesh, rank=2)

A[0, 0] = C[1, 1] = lame + 2 * shear_modulus
A[1, 1] = C[0, 0] = shear_modulus

B[0, 1] = D[1, 0] = lame
B[1, 0] = D[0, 1] = shear_modulus

#B[1, 0] = D[0, 1] = lame
#B[0, 1] = D[1, 0] = shear_modulus
```

```python
fixedPoint = fp.CellVariable(mesh=mesh)
fixedPoint[0] = 1e+10

eq_x = 0 == fp.DiffusionTerm(A, var=u_x) \
       + fp.DiffusionTerm(B, var=u_y) - f_x #- fp.ImplicitSourceTerm(fixedPoint, var=u_x)
eq_y = 0 == fp.DiffusionTerm(C, var=u_y) \
       + fp.DiffusionTerm(D, var=u_x) - f_y #- fp.ImplicitSourceTerm(fixedPoint, var=u_y)

eq = eq_x & eq_y
```

```python
from fipy.solvers.trilinos.linearGMRESSolver import LinearGMRESSolver

solver = LinearGMRESSolver()
#solver = fp.LinearLUSolver()
```

```python
u_x[:] = 0.0
u_y[:] = 0.0
res = eq.solve()
```
